```python
# classificar um triangulo
lado1 = float(input("digite a medida do lado 1: "))
lado2 = float(input("digite a medida do lado 2: "))
lado3 = float(input("digite a medida do lado 3: "))
if lado1 == lado2 and lado2 == lado3 and lado1 == lado3:
  print("E um triangulo equilatero3.")
elif lado1 != lado2 and lado2 != lado3 and lado1 != lado3:
  print("E um triangulo escaleno.")
else:
  print("E um triangulo isosceles.")
```

```python
# grafico linear
import numpy as np 
import matplotlib.pyplot as plt
x = np.linspace(0, 9, 10)
y = 2 * x
plt.plot(x, y, "b-")
plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt
x = np.linspace(0, 4 * np.pi, 1000)
y = np.sin(x)
plt.plot(x, y, "r--")
plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt
x = np.linspace(0, 20, 21)
for line_width in [0.5, 1, 2, 4, 8]:
  y = line_width * x 
  plt.plot(x, y, "k", linewidth=line_width)
plt.title("line width")
plt.grid(True)
plt.show()
```
